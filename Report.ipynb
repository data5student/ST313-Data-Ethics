### <b> 1. Introduction <a name="intro"></a> </b>

<b> Motivation and Ethical Context <a name="mot"></a> </b>

<p>
Contemporary health research is becoming increasingly reliant on observational studies, moving away from the ‘golden standard’ of using randomised controlled trials (RCT) <a href= "https://pmc.ncbi.nlm.nih.gov/articles/PMC7989480/">(Bajwa et al. 2021)</a> . As noted by Bajwa et al. it is not always feasible to conduct an RCT, especially in studies involving long-term dietary and lifestyle behaviours in which researchers need to resort to other experiment designs. These findings carry substantial weight as they shape public behaviour and health policy, as seen from the 1998 Wakefield MMR-autism study, where a single flawed study sparked an international anti-vaccination movement, leading to increases in preventable disease outbreaks (source). (or maybe something to do with surrogate science idk). Whilst some cases involve intentional fraud, even well-intentioned observational research can have misleading results due to choices in the method, causal assumptions, problems with the data, etc.
</p>
<p>
Li et al’s paper studying the relationship between coffee consumption and cognitive performance based on the National Health and Nutrition Examination Survey (NHANES 2011-2014) conducted by the CDC, effectively showcases the necessity and challenges of making observational research fair, accurate, and replicable. 
</p>

<h3 id="research-question">1.2 Research Question and Objectives</h3>
<p>
  The core aim of this study is to replicate and stress-test a published observational analysis linking caffeine intake
  to cognitive outcomes, using NHANES survey data. Rather than proposing a new algorithm, the focus is methodological:
  to evaluate how much the reported association depends on plausible alternative modelling choices and causal assumptions.
</p>

<p><b>Research question.</b> To what extent are estimated associations between caffeine intake and cognitive performance
  stable under different defensible covariate specifications, and what are the ethical implications of that sensitivity
  for interpreting observational results?
</p>

<p><b>Objectives.</b></p>
<ul>
  <li>
    Replicate the primary analysis using comparable variable definitions and regression specifications.
  </li>
  <li>
    Compare results across multiple cognitive outcomes to assess whether patterns are consistent or outcome-specific.
  </li>
  <li>
    Conduct sensitivity analysis by fitting alternative models that reflect different causal roles of key variables
    (e.g., treating some variables as confounders versus mediators).
  </li>
  <li>
    Interpret findings through an ethics lens, focusing on transparency, researcher degrees of freedom, and the risk of
    overclaiming causality from observational associations.
  </li>
</ul>
