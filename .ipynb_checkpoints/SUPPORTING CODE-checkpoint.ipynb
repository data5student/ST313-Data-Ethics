### <b> 1. Model Introduction <a name="model"></a> </b>

<b> Overview </b>

Below is the code for all the models. For each, we use the library <b> daggity </b> to create the dag graphs. In the final report, we use our own graphs of models for neatness due to the limitations for the daggity graphs. There are three outcome variables, but instead of having three identical graphs for each model with different labels, we represent <b> CERAD, DSST, </b> and <b> Animal Fluency </b> tests with the label <b> CognitivePerformance </b>.

We have four models, the first three which are replicated from the original study and the fourth is our own proposed model. In the study, they first use an unadjusted, naive model with simply caffiene intake --> cognitive performance. The second model adjusts for sex and age, the third adjusts for more covariates (Alchol Intake, BMI, Marital Status, Stroke, Race, Diabetes, and Smoking). The fourth is our own proposed model through which we aim to show that there is no one true causual graph and that the relationships you assume as well as what you label as mediators, colliders, covariates etc. does in fact affect the outcome of the result.

To replicate the paper as close as we can, we use similar techniques for all models. First we apply linear regression to estimate causal effect of increasing caffiene consumption by 50 mg daily for each cognitive performance outcome. Then through logistic regression we calculate the odds ratio, by labelling the bottom 25% as low cognitive performance or cognitive impairment and comparing with the fourth quartile. We also calculate the p-value and significance so that we can compare our results with the results listed in the paper. 


<b> Model Sequence </b> 

We fit four nested models to progressively account and adjust for potential confounding and mediating variables. This progression allows us to analyse how estimated effects of caffiene intake change as some variables are reclassified from confounder to mediators or excluded from having no effect. We display the outcomes from the first model, first outcome variable (CERAD) for simplicity sake. The rest of variables are stroed and displayed in a table later on when needed for analysis. 

<b> Model 1 (Unadjusted / Naive) </b> 
This only includes caffeine intake and so estimates the unadjusted association between caffeine intake and cognitive performance.

<b> Model 2 (Adjusted for Sex and Age) </b>
We adjust for age and sex, treating these variables as baseline confounders affecting both caffeine consumption and cognitive performance.

<b> Model 3 (Adjusted for Further Covariates)</b> 
We further adjust for race/ethnicity, marital status, BMI, smoking, alcohol intake, diabetes, and stroke, which corresponds to the main multivariable adjustment that the original study reports.

<b> Model 4 (Our Proposed Alternate Model) </b>
Our model keeps age, sex, race/ethnicity, smoking status, and BMI, and excludes variables such marital status, alcohol intake, diabetes, and stroke. This reflects our causal interpretation in which certain variables are treated as downstream mediators or colliders rather than confounders.

### <b> 2. Model 1 (Unadjusted / Naive) <a name="m1"></a> </b>

```python
%%R
library(dagitty)

dag_m1 <- dagitty("dag {Caffeine -> CognitivePerformance }")
plot(dag_m1)
```

<b> Model 1a: CERAD delayed recall </b>

```python
%%R
# linear regression CERAD - Caffiene
m1a_data <- subset(df, !is.na(CFDCSR) & !is.na(DRXTCAFF))
model_1a <- lm(CFDCSR ~ DRXTCAFF, data = m1a_data)
summary(model_1a)
```

```python
%%R
# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m1a_data$DRXTCAFF)

pred_plus50 <- predict(model_1a, newdata = data.frame(DRXTCAFF = mean_caff + 50))
pred_base <- predict(model_1a, newdata = data.frame(DRXTCAFF = mean_caff))
effect_plus50_1a <- pred_plus50 - pred_base
effect_plus50_1a
```

```python
%%R
# creating caffiene quartiles
df$caff_q <- cut(
  df$DRXTCAFF,
  breaks = quantile(df$DRXTCAFF, probs = c(0, .25, .5, .75, 1), na.rm = TRUE),
  include.lowest = TRUE,
  labels = c("Q1", "Q2", "Q3", "Q4"))

df$caff_q <- relevel(df$caff_q, ref = "Q1")

cerad_cut <- quantile(df$CFDCSR, 0.25, na.rm = TRUE)
df$Y_lowCERAD <- ifelse(df$CFDCSR <= cerad_cut, 1, 0)

table(df$caff_q, useNA = "ifany")
```

```python
%%R
# logistic regression in quartiles to compare with the original paper
m1_cerad <- subset(df, !is.na(Y_lowCERAD) & !is.na(caff_q))

fit_m1_cerad <- glm(Y_lowCERAD ~ caff_q, family = binomial(), data = m1_cerad)
summary(fit_m1_cerad)
```

```python
%%R
# odds ratio
OR_m1 <- exp(cbind(OR = coef(fit_m1_cerad), confint(fit_m1_cerad)))
OR_m1
```

```python
%%R
# saving odds ratio from fourth quartile
OR_1a <- OR_m1["caff_qQ4", "OR"]

# saving p-value from fourt quartile
p_1a <- summary(fit_m1_cerad)$coefficients["caff_qQ4", "Pr(>|z|)"]

# function to so that significance matches the same style as the original report
p_to_sigcat <- function(p) {
  if (is.na(p)) return(NA)
  if (p < 0.001) return("< 0.001")
  if (p < 0.01)  return("< 0.01")
  if (p < 0.05)  return("< 0.05")
  return("≥ 0.05")
}

sig_1a <- sapply(p_1a, p_to_sigcat)
sig_1a
```

<b> Model 1b: Digit Symbol Substitution Test </b>
Code repeated from above

```python
%%R
# linear regression DSST - Caffiene
m1b_data <- subset(df, !is.na(CFDDS) & !is.na(DRXTCAFF))
model_1b <- lm(CFDDS ~ DRXTCAFF, data = m1b_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m1b_data$DRXTCAFF)

pred_plus50 <- predict(model_1b, newdata = data.frame(DRXTCAFF = mean_caff + 50))
pred_base <- predict(model_1b, newdata = data.frame(DRXTCAFF = mean_caff))

effect_plus50_1b <- pred_plus50 - pred_base

# creating DSS  outcome
dsst_cut <- quantile(df$CFDDS, 0.25, na.rm = TRUE)
df$Y_lowDSST <- ifelse(df$CFDDS <= dsst_cut, 1, 0)

# logistic regression in quartiles to compare with the original paper
m1_dsst <- subset(df, !is.na(Y_lowDSST) & !is.na(caff_q))
fit_m1_dsst <- glm(Y_lowDSST ~ caff_q, family = binomial(), data = m1_dsst)

# odds ratio
OR_m1b <- exp(cbind(OR = coef(fit_m1_dsst), confint(fit_m1_dsst)))

# saving odds ratio from fourth quartile
OR_1b <- OR_m1b["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_1b <- summary(fit_m1_dsst)$coefficients["caff_qQ4", "Pr(>|z|)"]
```

<b> Model 1c: Animal Fluency Test </b>

```python
%%R
# linear regression Animal Fluency - Caffiene
m1c_data <- subset(df, !is.na(CFDAST) & !is.na(DRXTCAFF))
model_1c <- lm(CFDAST ~ DRXTCAFF, data = m1c_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m1c_data$DRXTCAFF)

pred_plus50 <- predict(model_1c, newdata = data.frame(DRXTCAFF = mean_caff + 50))
pred_base <- predict(model_1c, newdata = data.frame(DRXTCAFF = mean_caff))

effect_plus50_1c <- pred_plus50 - pred_base

# creating Animal Fluency outcome
af_cut <- quantile(df$CFDAST, 0.25, na.rm = TRUE)
df$Y_lowAF <- ifelse(df$CFDAST <= af_cut, 1, 0)

# logistic regression in quartiles to compare with the original paper
m1_af <- subset(df, !is.na(Y_lowAF) & !is.na(caff_q))
fit_m1_af <- glm(Y_lowAF ~ caff_q, family = binomial(), data = m1_af)

# odds ratio
OR_m1c <- exp(cbind(OR = coef(fit_m1_af), confint(fit_m1_af)))

# saving odds ratio from fourth quartile
OR_1c <- OR_m1c["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_1c <- summary(fit_m1_af)$coefficients["caff_qQ4", "Pr(>|z|)"]
```

```python
%%R
# summary of model 1

model1_results <- data.frame(
  Outcome = c("CERAD delayed recall", "DSST", "Animal Fluency"),
  OR_Q4_vs_Q1 = c(OR_1a, OR_1b,OR_1c),
  P_value_Q4_vs_Q1 = c(p_1a, p_1b, p_1c),
  Sig_1 = sapply(c(p_1a, p_1b, p_1c), p_to_sigcat),
  Effect_plus_50mg = c(effect_plus50_1a, effect_plus50_1b, effect_plus50_1c))

model1_results
```

<b> Model 1 results </b>

A 50 mg/day increase in caffiene intake is associated with a 0.08 point increase in the CERAD delayed recall score, 0.76 points in DSST, and 0.30 for Animal Fluency score. Overall there is weak positive unadjusted associations between caffiene intake and the cognitive test scores, however these should not be taken as causal effect as other health and lifestyle factors are potential confounders in the relationship, which is what the subsequent models should progressively adjust for.

### <b> 5. Model 4 (Our Proposed Alternate Model) <a name="m4"></a> </b>

In our proposed model, we deliberately change the adjustment set from the previous models to highlight our point on causal interence; there is no one true regression case and you can commit to any causal inference graph within a variety of options. Instead of using statistical checks which can demonstrate associations, we primarly use causal structure logic to create our DAG Model. In our case, we believe that age and sex are important <b> confounders </b> of the caffiene and cognitive performance association as they likely affect both. We think that marital status and alcohol intake aren't valid confounders. We also adjust for race as a baseline <b> covariate </b> because we think it may confound social or behavioural patterns linked to our model. Confounders must be adjusted for to remove bias, however if we adjust for mediators we may block part of the total effect and if we adjust for colliders, we could create spurious associations.

The main difference is that we believe both BMI and smoking as potential <b> mediators </b> because they likely lie on the pathway from caffiene intake to cognitive function. Increased caffiene intake could imply less concern for health so a higher BMI and more cigarettes smoked. By adjusting for these, the previous models may be blocking part of the previous effect so in this model we don't adjust for these mediators.

Similarly, we think both diabetes and stroke should be treated as <b> colliders </b>. Using the logic that a higher caffiene intake may mean higher chances of diabetes or stroke and that these chances are also affected by age, BMI, and smoking. So if we treat diabetes and stroke as collider variables, we shouldn't adjust for them to avoid opening non-causal paths and inducing bias. 

With all these changes, our proposed model demonstrated how causal effect estimates can change when we reclassify variables using different plausible causal assumption. The purpose is to highlight the importance of explicit causal reasoning in observational analyses and to be transparent in all our in-processing decisions.

```python
%%R
# 4a CERAD

# linear regression CERAD - Caffiene
m4a_data <- subset(df, !is.na(CFDCSR) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

model_4a <- lm(CFDCSR ~ DRXTCAFF + RIDAGEYR + RIAGENDR + RIDRETH1, data = m4a_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m4a_data$DRXTCAFF)

pred_plus50 <- predict(model_4a, newdata = transform(m4a_data[1, ], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_4a, newdata = transform(m4a_data[1, ], DRXTCAFF = mean_caff))

effect_plus50_4a <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m4_cerad <- subset(df, !is.na(Y_lowCERAD) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

fit_m4_cerad <- glm(Y_lowCERAD ~ caff_q + RIDAGEYR + RIAGENDR + RIDRETH1, family = binomial(), data = m4_cerad)

# odds ratio
OR_m4a <- exp(cbind(OR = coef(fit_m4_cerad), confint(fit_m4_cerad)))

# saving odds ratio from fourth quartile
OR_4a <- OR_m4a["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_4a <- summary(fit_m4_cerad)$coefficients["caff_qQ4", "Pr(>|z|)"]


# 4b DSST

# linear regression DSST - Caffiene
m4b_data <- subset(df, !is.na(CFDDS) & !is.na(DRXTCAFF) &!is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

model_4b <- lm(CFDDS ~ DRXTCAFF + RIDAGEYR + RIAGENDR + RIDRETH1, data = m4b_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m4b_data$DRXTCAFF)

pred_plus50 <- predict(model_4b, newdata = transform(m4b_data[1, ], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_4b, newdata = transform(m4b_data[1, ], DRXTCAFF = mean_caff))

effect_plus50_4b <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m4_dsst <- subset(df, !is.na(Y_lowDSST) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

fit_m4_dsst <- glm(Y_lowDSST ~ caff_q + RIDAGEYR + RIAGENDR + RIDRETH1, family = binomial(), data = m4_dsst)

# odds ratio
OR_m4b <- exp(cbind(OR = coef(fit_m4_dsst), confint(fit_m4_dsst)))

# saving odds ratio from fourth quartile
OR_4b <- OR_m4b["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_4b <- summary(fit_m4_dsst)$coefficients["caff_qQ4", "Pr(>|z|)"]


# 4c Animal Fluency

# linear regression Animal Fluency - Caffiene
m4c_data <- subset(df, !is.na(CFDAST) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

model_4c <- lm(CFDAST ~ DRXTCAFF + RIDAGEYR + RIAGENDR + RIDRETH1, data = m4c_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m4c_data$DRXTCAFF)

pred_plus50 <- predict(model_4c, newdata = transform(m4c_data[1, ], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_4c, newdata = transform(m4c_data[1, ], DRXTCAFF = mean_caff))

effect_plus50_4c <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m4_af <- subset(df, !is.na(Y_lowAF) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

fit_m4_af <- glm(Y_lowAF ~ caff_q + RIDAGEYR + RIAGENDR + RIDRETH1, family = binomial(), data = m4_af)

# odds ratio
OR_m4c <- exp(cbind(OR = coef(fit_m4_af), confint(fit_m4_af)))

# saving odds ratio from fourth quartile
OR_4c <- OR_m4c["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_4c <- summary(fit_m4_af)$coefficients["caff_qQ4", "Pr(>|z|)"]
```

### <b> 8. Mediator Total Effect Breakdown Calculation <a name="effect"></a> </b>

Most of the code in this section is adapted from the Week 7 Class code.

Total effect is the sum of natural direct effect <b> (NDE) </b> and natural indirect effect <b> (NIE)</b>. The NDE measures how much the cognitive function test results would change if caffiene intake increased and the mediator was kept fixed at the level it would naturally be at without the extra caffiene. The NIE measures how much the test results would change due to only the caffiene changing the mediator. Both of these in tandem help us to recognise the effect of a specific variable and can help provide further clarification when classifying as a non-confounder. 

So we calculate these effects for both BMI and Smoking, variables we believe to be mediators. 



#### First for BMI with Outcome Variable DSST

For the first model, outcome is DSST Score and the mediator is BMI. Later, I will repeat the code for the outcomes CERAD test score and Animal Fluency score, and then run the combination for mediator as smoking.

In this case, A is caffiene intake where A0 is the mean and A1 is A0 + 50mg.
Y is the outcome, the indicator for a low DSST score where we define low to be the bottom quartile.
X is the mediator BMI.
Giving the Causal diagram of A --> X --> Y.

Let Y_a be the outcome we would observe if caffiene was set to A = a
Let X_a be the mediator value we would observe if caffiene was set to A = a
Let Y_ax be the outcome if caffiene were set to A = a and the mediator was forced to be x. 

```python
%%R
# sorting to complete cases
dat_bmi <- subset(df, !is.na(Y_lowDSST) & !is.na(DRXTCAFF) & !is.na(BMXBMI) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

# defining A0 and A1
A0 <- mean(dat_bmi$DRXTCAFF, na.rm = TRUE)
A1 <- A0 + 50
cat(A0, A1)
```

##### Total Effect
The total effect compares the expected risk if caffiene was to increase by 50 mg to the average. 

Where TE = E[Y_A1] - E[Y_A0].

The code below gives E[Y_A1] = 0.233. E[Y_A0] = 0.256. With the total effect being -0.023. Implying that by increasing caffiene by 50mg a day, there is a 2.3 percentage point lower risk of low DSST test performance.

```python
%%R

set.seed(2025)

# the model (M4 with bmi)
outcome_bmi <- glm(Y_lowDSST ~ DRXTCAFF + BMXBMI + RIDAGEYR + RIAGENDR + RIDRETH1, data = dat_bmi, family = binomial)

# E[Y_A0]
p_Y_A0 <- predict(outcome_bmi,newdata = transform(dat_bmi, DRXTCAFF = A0), type = "response")

# E[Y_A1]
p_Y_A1 <- predict(outcome_bmi, newdata = transform(dat_bmi, DRXTCAFF = A1), type = "response")

TE <- mean(p_Y_A1) - mean(p_Y_A0)
cat(mean(p_Y_A0), mean(p_Y_A1), TE)
```

##### Controlled Direct Effect
To calculate the controlled direct effect, we have to fix X to a constant value, which we use the sample mean for BMI. This allows us to isolate the direct effect of caffiene. 

CDE(x) = E[Y_A1_x] - E[Y_A0_x]

E[Y_A1_X] IS 0.256. E[Y_A0_X] = 0.233.
CDE = -0.0227. 

```python
%%R
# using the average to fix
bmi_fixed <- mean(dat_bmi$BMXBMI, na.rm = TRUE)

# E[Y_A0_x] 
p_Y_A0_x <- predict(outcome_bmi, newdata = transform(dat_bmi, DRXTCAFF = A0, BMXBMI  = bmi_fixed), type = "response")

# E[Y_A1_x]
p_Y_A1_x <- predict(outcome_bmi, newdata = transform(dat_bmi, DRXTCAFF = A1, BMXBMI  = bmi_fixed), type = "response")

CDE <- mean(p_Y_A1_x) - mean(p_Y_A0_x)
cat(mean(p_Y_A0_x), mean(p_Y_A1_x), CDE)
```

##### Natural Direct Effect
The Natural Direct Effect compares the outcome when caffiene changes but BMI is held at the natural value when A = A0 to give X_A0.

NDE = E[Y_{A1, X_A0}] − E[Y_{A0,X_A0}]

E[Y_{A1, X_A0}]  = 0.233
E[Y_{A0,X_A0}] = 0.256
NDE = -0.023

```python
%%R

set.seed(2026)

bmi_model <- lm(BMXBMI ~ DRXTCAFF + RIDAGEYR + RIAGENDR + RIDRETH1, data = dat_bmi)

# X_A0
mu_X_A0 <- predict(bmi_model, newdata = transform(dat_bmi, DRXTCAFF = A0))
sd_X <- summary(bmi_model)$sigma
X_A0 <- rnorm(nrow(dat_bmi), mean = mu_X_A0, sd = sd_X)

# E[Y_{A0,X_A0}]
p_Y_A0_XA0 <- predict(outcome_bmi, newdata = transform(dat_bmi, DRXTCAFF = A0, BMXBMI  = X_A0), type = "response")

# E[Y_{A1, X_A0}]
p_Y_A1_XA0 <- predict(outcome_bmi, newdata = transform(dat_bmi, DRXTCAFF = A1, BMXBMI  = X_A0), type = "response")

NDE <- mean(p_Y_A1_XA0) - mean(p_Y_A0_XA0)
cat(mean(p_Y_A0_XA0), mean(p_Y_A1_XA0), NDE)
```

##### Natural Indirect Effect
The natural indirect efect fixes caffiene at A = A0 and this time changes BMI to compare how the output would change when BMI shifts from using X_A0 to X_A1.

NIE=E[Y_{A0, X_A1}] − E[Y_{A0,X_A0}]

NIE = -0.0013, which is a very miniscule value which implies that the part of caffiene's effect on the DSST score that is mediated through changes in BMI is essentialy zero. 

```python
%%R
set.seed(2027)

# X_{A1}
mu_X_A1 <- predict(bmi_model, newdata = transform(dat_bmi, DRXTCAFF = A1))
X_A1 <- rnorm(nrow(dat_bmi), mean = mu_X_A1, sd = sd_X)

# E[Y_{A0, X_A1}]
p_Y_A0_XA1 <- predict(outcome_bmi, newdata = transform(dat_bmi, DRXTCAFF = A0, BMXBMI  = X_A1), type = "response")

NIE <- mean(p_Y_A0_XA1) - mean(p_Y_A0_XA0)
cat(mean(p_Y_A0_XA0), mean(p_Y_A0_XA1), NIE)
```

##### The Decomposition Identity
The decomposition identity says that total effect can be written as the sum of natural indirect and direct effect. 
TE = NDE + NIE

However, as seen from the resutls below, in code we only use estimates so it doesn't always match we up exactly. We can see it is simlar enough which lets us know that we computed the values correctly. 


```python
%%R
cat(TE, NDE, NIE, NDE + NIE)
```

Overall, this code shows us that BMI is not a mediator in the caffiene and DSST relationship, in contrast to what we thought and assumed it was in our model 4. This goes to show that our causal reasoning in incorrect.
```python
%%R
#BMI - ANIMAL FLUENCY

# sorting to complete cases
dat_bmi_af <- subset(df, !is.na(Y_lowAF) & !is.na(DRXTCAFF) & !is.na(BMXBMI) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

# defining A0 and A1
A0_af <- mean(dat_bmi_af$DRXTCAFF, na.rm = TRUE); A1_af <- A0_af + 50

set.seed(2025)

# the model (M4 with bmi)
outcome_bmi_af <- glm(Y_lowAF ~ DRXTCAFF + BMXBMI + RIDAGEYR + RIAGENDR + RIDRETH1, data = dat_bmi_af, family = binomial)

# total effect
p_Y_A0_af <- predict(outcome_bmi_af, newdata = transform(dat_bmi_af, DRXTCAFF = A0_af), type = "response")
p_Y_A1_af <- predict(outcome_bmi_af, newdata = transform(dat_bmi_af, DRXTCAFF = A1_af), type = "response")
TE_af_bmi <- mean(p_Y_A1_af) - mean(p_Y_A0_af)

# controlled direct effect
bmi_fixed_af <- mean(dat_bmi_af$BMXBMI, na.rm = TRUE)
p_Y_A0_x_af <- predict(outcome_bmi_af, newdata = transform(dat_bmi_af, DRXTCAFF = A0_af, BMXBMI = bmi_fixed_af), type = "response")
p_Y_A1_x_af <- predict(outcome_bmi_af, newdata = transform(dat_bmi_af, DRXTCAFF = A1_af, BMXBMI = bmi_fixed_af), type = "response")
CDE_af_bmi <- mean(p_Y_A1_x_af) - mean(p_Y_A0_x_af)

set.seed(2026)

# natural direct effect
bmi_model_af <- lm(BMXBMI ~ DRXTCAFF + RIDAGEYR + RIAGENDR + RIDRETH1, data = dat_bmi_af)
mu_X_A0_af <- predict(bmi_model_af, newdata = transform(dat_bmi_af, DRXTCAFF = A0_af))
sd_X_af <- summary(bmi_model_af)$sigma
X_A0_af <- rnorm(nrow(dat_bmi_af), mean = mu_X_A0_af, sd = sd_X_af)
p_Y_A0_XA0_af <- predict(outcome_bmi_af, newdata = transform(dat_bmi_af, DRXTCAFF = A0_af, BMXBMI = X_A0_af), type = "response")
p_Y_A1_XA0_af <- predict(outcome_bmi_af, newdata = transform(dat_bmi_af, DRXTCAFF = A1_af, BMXBMI = X_A0_af), type = "response")
NDE_af_bmi <- mean(p_Y_A1_XA0_af) - mean(p_Y_A0_XA0_af)

set.seed(2027)

# natural indirect effect
mu_X_A1_af <- predict(bmi_model_af, newdata = transform(dat_bmi_af, DRXTCAFF = A1_af))
X_A1_af <- rnorm(nrow(dat_bmi_af), mean = mu_X_A1_af, sd = sd_X_af)
p_Y_A0_XA1_af <- predict(outcome_bmi_af, newdata = transform(dat_bmi_af, DRXTCAFF = A0_af, BMXBMI = X_A1_af), type = "response")
NIE_af_bmi <- mean(p_Y_A0_XA1_af) - mean(p_Y_A0_XA0_af)
```

```python
%%R

# SMOKE TO DSST

# recoding smoking
smoke_bin <- ifelse(df$SMQ020 == 1, 1, ifelse(df$SMQ020 == 2, 0, NA))

# sorting to complete cases
dat_smk_dsst <- subset(df, !is.na(Y_lowDSST) & !is.na(DRXTCAFF) & !is.na(smoke_bin) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(RIDRETH1))

# defining A0 and A1
A0_smk_dsst <- mean(dat_smk_dsst$DRXTCAFF, na.rm = TRUE); A1_smk_dsst <- A0_smk_dsst + 50

set.seed(2025)

# the model (M4 with smoking)
outcome_smk_dsst <- glm(Y_lowDSST ~ DRXTCAFF + smoke_bin + RIDAGEYR + RIAGENDR + RIDRETH1, data = dat_smk_dsst, family = binomial)

# E[Y_A0]
p_Y_A0_smk_dsst <- predict(outcome_smk_dsst, newdata = transform(dat_smk_dsst, DRXTCAFF = A0_smk_dsst), type = "response")

# E[Y_A1]
p_Y_A1_smk_dsst <- predict(outcome_smk_dsst, newdata = transform(dat_smk_dsst, DRXTCAFF = A1_smk_dsst), type = "response")

TE_dsst_smk <- mean(p_Y_A1_smk_dsst) - mean(p_Y_A0_smk_dsst)

set.seed(2026)

smk_model_dsst <- glm(smoke_bin ~ DRXTCAFF + RIDAGEYR + RIAGENDR + RIDRETH1, data = dat_smk_dsst, family = binomial)

# X_A0
X_A0_smk_dsst <- rbinom(nrow(dat_smk_dsst), 1, predict(smk_model_dsst, newdata = transform(dat_smk_dsst, DRXTCAFF = A0_smk_dsst), type = "response"))

# E[Y_{A0,X_A0}]
p_Y_A0_XA0_smk_dsst <- predict(outcome_smk_dsst, newdata = transform(dat_smk_dsst, DRXTCAFF = A0_smk_dsst, smoke_bin = X_A0_smk_dsst), type = "response")

# E[Y_{A1, X_A0}]
p_Y_A1_XA0_smk_dsst <- predict(outcome_smk_dsst, newdata = transform(dat_smk_dsst, DRXTCAFF = A1_smk_dsst, smoke_bin = X_A0_smk_dsst), type = "response")

NDE_dsst_smk <- mean(p_Y_A1_XA0_smk_dsst) - mean(p_Y_A0_XA0_smk_dsst)

set.seed(2027)

# X_A1
X_A1_smk_dsst <- rbinom(nrow(dat_smk_dsst), 1, predict(smk_model_dsst, newdata = transform(dat_smk_dsst, DRXTCAFF = A1_smk_dsst), type = "response"))

# E[Y_{A0, X_A1}]
p_Y_A0_XA1_smk_dsst <- predict(outcome_smk_dsst, newdata = transform(dat_smk_dsst, DRXTCAFF = A0_smk_dsst, smoke_bin = X_A1_smk_dsst), type = "response")

NIE_dsst_smk <- mean(p_Y_A0_XA1_smk_dsst) - mean(p_Y_A0_XA0_smk_dsst)
```

For BMI, the majority of the total effect is explained by the natural direct effect in all three outcome cases, where the NIE is a few orders of magnititude smaller than the total effect, making it virtually negligible. This implies that BMI is not really a meaningful mediator betwee caffiene and cognitive function. Our Model 4 might be improved if we interpret BMI as a covariate or confounder isntead of a mediator, and by mislabelling it, we may have introduced some bias in our model. The relatively small total effect value indicates that BMI is not a particularly strong confounder either. 

Similarly for smoking, the natural direct effect is far greater than indirect, but looking specifcally at CERAD, NIE is 16% of the total TE. So it does explaing a small fraction of cafiene's effect on CERAD test score. Classification here is a little bit ambiguous, but at best smoking is a very weak mediator and still not a primary causal pathway. 

Both Smoking and BMI, for which we took the causal assumption and classified them as mediators, may have been misclassed in our models. Treating these variables as mediators in model 4 was largely unnecessary and it may have even been inaproppriate for our defined relationship.
