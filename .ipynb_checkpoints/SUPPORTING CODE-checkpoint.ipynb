Model Overview and Analytical Strategy

In this section, we outline the sequence of statistical models used to examine the association between caffeine intake and cognitive performance, following the modelling strategy of the original study and the causal framework introduced in ST313 (Classes 6 and 7).

Cognitive performance was assessed using three outcomes:
(i) CERAD delayed recall score,
(ii) Digit Symbol Substitution Test (DSST) score, and
(iii) Animal Fluency score.

Each outcome was analysed separately, and all models were fitted independently for each cognitive test. For brevity, the main report presents one representative model per stage (Model a), while the supporting code contains the full set of models (a–c) for all outcomes.

Caffeine Exposure

Daily caffeine intake (DRXTCAFF, mg/day) was analysed in two ways:

Continuous exposure, to estimate the average causal effect of a +50 mg/day increase in caffeine intake using linear and logistic regression models.

Categorical exposure, where caffeine intake was divided into quartiles, with the lowest quartile serving as the reference group, to allow comparison with the odds ratios reported in the original study.

Binary Cognitive Impairment Outcomes

To align with the original paper, cognitive impairment was defined separately for each test as a score below the 25th percentile of the distribution within the analytic sample. This resulted in three binary outcomes:

CERAD impairment

DSST impairment

Animal Fluency impairment

Participants were included in each analysis only if data for the corresponding cognitive test were available.

Model Sequence

Four nested models were fitted to progressively adjust for potential confounding and mediating variables:

Model 1 (Crude model)
Includes caffeine intake only.
This model estimates the unadjusted association between caffeine intake and cognitive performance.

Model 2 (Demographic-adjusted model)
Adjusts for age and sex.
These variables are treated as baseline confounders affecting both caffeine consumption and cognitive performance.

Model 3 (Fully adjusted confounding model)
Further adjusts for race/ethnicity, marital status, body mass index (BMI), smoking status, alcohol consumption, diabetes, and stroke.
This model corresponds to the main multivariable adjustment reported in the original study.

Model 4 (Causal pathway model)
Retains age, sex, race/ethnicity, smoking status, and BMI, while excluding variables such as marital status, alcohol intake, diabetes, and stroke.
This model reflects a causal interpretation in which certain variables are treated as downstream mediators or colliders rather than confounders, consistent with the directed acyclic graphs (DAGs) presented in the main report.

Causal Interpretation

The progression from Model 1 to Model 4 allows us to examine how estimated effects of caffeine intake change as variables are reclassified from confounders to mediators or excluded altogether. In addition to regression-based estimates, causal effect decompositions—such as the Controlled Direct Effect (CDE), Natural Direct Effect (NDE), and Natural Indirect Effect (NIE)—are computed for selected mediators (e.g., BMI) to assess whether the effect of caffeine on cognitive outcomes operates through these pathways.

This modelling strategy enables a direct comparison with the original paper’s results while also illustrating key causal concepts from ST313, including confounding adjustment, mediation, and the distinction between association and causal effect.

### <b> 2. Model 1 (Unadjusted / Naive) <a name="m1"></a> </b>
We first estimated the unadjusted association between caffeine intake and cognitive impairment using logistic regression. Cognitive impairment was defined as performance below the 25th percentile for each cognitive test. Separate models were fitted for the CERAD delayed recall test, the Digit Symbol Substitution Test (DSST), and the Animal Fluency test.

Following the paper, Model 1 is unadjusted: we regress cognitive performance on caffeine intake only. We show one causal DAG for intuition (Age and Sex are plausible confounders), but we do not adjust for them in Model 1, to match the paper.
We then illustrate a simple model-based “+50 mg/day caffeine” contrast by predicting outcomes at two caffeine levels and taking the difference. This should be interpreted as an unadjusted association, not a definitive causal effect.

```python
%%R
library(dagitty)

dag_m1 <- dagitty("dag {Caffeine -> CognitivePerformance }")
plot(dag_m1)
```

<b> Model 1a: CERAD delayed recall </b>

```python
%%R
# linear regression CERAD - Caffiene
m1a_data <- subset(df, !is.na(CFDCSR) & !is.na(DRXTCAFF))
model_1a <- lm(CFDCSR ~ DRXTCAFF, data = m1a_data)
summary(model_1a)
```

```python
%%R
# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m1a_data$DRXTCAFF)

pred_plus50 <- predict(model_1a, newdata = data.frame(DRXTCAFF = mean_caff + 50))
pred_base <- predict(model_1a, newdata = data.frame(DRXTCAFF = mean_caff))
effect_plus50_1a <- pred_plus50 - pred_base
effect_plus50_1a
```

```python
%%R
# creating caffiene quartiles
df$caff_q <- cut(
  df$DRXTCAFF,
  breaks = quantile(df$DRXTCAFF, probs = c(0, .25, .5, .75, 1), na.rm = TRUE),
  include.lowest = TRUE,
  labels = c("Q1", "Q2", "Q3", "Q4"))

df$caff_q <- relevel(df$caff_q, ref = "Q1")

cerad_cut <- quantile(df$CFDCSR, 0.25, na.rm = TRUE)
df$Y_lowCERAD <- ifelse(df$CFDCSR <= cerad_cut, 1, 0)

table(df$caff_q, useNA = "ifany")
```

```python
%%R
# logistic regression in quartiles to compare with the original paper
m1_cerad <- subset(df, !is.na(Y_lowCERAD) & !is.na(caff_q))

fit_m1_cerad <- glm(Y_lowCERAD ~ caff_q, family = binomial(), data = m1_cerad)
summary(fit_m1_cerad)
```

```python
%%R
# odds ratio
OR_m1 <- exp(cbind(OR = coef(fit_m1_cerad), confint(fit_m1_cerad)))
OR_m1
```

```python
%%R
# saving odds ratio from fourth quartile
OR_1a <- OR_m1["caff_qQ4", "OR"]

# saving p-value from fourt quartile
p_1a <- summary(fit_m1_cerad)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_1a <- p_1a < 0.05
```

<b> Model 1b: Digit Symbol Substitution Test </b>
Code repeated from above

```python
%%R
# linear regression DSST - Caffiene
m1b_data <- subset(df, !is.na(CFDDS) & !is.na(DRXTCAFF))
model_1b <- lm(CFDDS ~ DRXTCAFF, data = m1b_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m1b_data$DRXTCAFF)

pred_plus50 <- predict(model_1b, newdata = data.frame(DRXTCAFF = mean_caff + 50))
pred_base <- predict(model_1b, newdata = data.frame(DRXTCAFF = mean_caff))

effect_plus50_1b <- pred_plus50 - pred_base

# creating DSS  outcome
dsst_cut <- quantile(df$CFDDS, 0.25, na.rm = TRUE)
df$Y_lowDSST <- ifelse(df$CFDDS <= dsst_cut, 1, 0)

# logistic regression in quartiles to compare with the original paper
m1_dsst <- subset(df, !is.na(Y_lowDSST) & !is.na(caff_q))
fit_m1_dsst <- glm(Y_lowDSST ~ caff_q, family = binomial(), data = m1_dsst)

# odds ratio
OR_m1b <- exp(cbind(OR = coef(fit_m1_dsst), confint(fit_m1_dsst)))

# saving odds ratio from fourth quartile
OR_1b <- OR_m1b["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_1b <- summary(fit_m1_dsst)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_1b <- p_1b < 0.05
```

<b> Model 1c: Animal Fluency Test </b>

```python
%%R
# linear regression Animal Fluency - Caffiene
m1c_data <- subset(df, !is.na(CFDAST) & !is.na(DRXTCAFF))
model_1c <- lm(CFDAST ~ DRXTCAFF, data = m1c_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m1c_data$DRXTCAFF)

pred_plus50 <- predict(model_1c, newdata = data.frame(DRXTCAFF = mean_caff + 50))
pred_base <- predict(model_1c, newdata = data.frame(DRXTCAFF = mean_caff))

effect_plus50_1c <- pred_plus50 - pred_base

# creating Animal Fluency outcome
af_cut <- quantile(df$CFDAST, 0.25, na.rm = TRUE)
df$Y_lowAF <- ifelse(df$CFDAST <= af_cut, 1, 0)

# logistic regression in quartiles to compare with the original paper
m1_af <- subset(df, !is.na(Y_lowAF) & !is.na(caff_q))
fit_m1_af <- glm(Y_lowAF ~ caff_q, family = binomial(), data = m1_af)

# odds ratio
OR_m1c <- exp(cbind(OR = coef(fit_m1_af), confint(fit_m1_af)))

# saving odds ratio from fourth quartile
OR_1c <- OR_m1c["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_1c <- summary(fit_m1_af)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_1c <- p_1c < 0.05
```

```python
%%R
# summary of model 1

model1_results <- data.frame(
  Outcome = c("CERAD delayed recall", "DSST", "Animal Fluency"),
  OR_Q4_vs_Q1 = c(OR_1a, OR_1b,OR_1c),
  P_value_Q4_vs_Q1 = c(p_1a, p_1b, p_1c),
  Significant_0.05 = c(sig_1a, sig_1b, sig_1c),
  Effect_plus_50mg = c(effect_plus50_1a, effect_plus50_1b, effect_plus50_1c))

model1_results
```

### <b> 3. Model 2 (Adjusted for Sex and Age) <a name="m2"></a> </b>
Model 2: Age- and sex-adjusted association between caffeine intake and cognitive performance

Model 2 adjusts the association between caffeine intake and cognitive performance for age and sex, as shown in Figure 2. These variables are treated as confounders because they may influence both caffeine consumption patterns and cognitive performance.

Separate linear regression models were estimated for each cognitive outcome. As in Model 1, we report a model-based contrast corresponding to a 50 mg/day increase in caffeine intake. These estimates represent adjusted associations under the assumed linear model and should not be interpreted as causal effects.

```python
%%R
dag_m2 <- dagitty("dag {

  Age -> Caffiene
  Sex -> Caffiene

  Age -> CognitivePerformance
  Sex -> CognitivePerformance

  Caffiene -> CognitivePerformance}")

plot(dag_m2)
```

```python
%%R
# Model 2

# a) CERAD

# linear regression CERAD - Caffiene
m2a_data <- subset(df, !is.na(CFDCSR) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR))
model_2a <- lm(CFDCSR ~ DRXTCAFF + RIDAGEYR + RIAGENDR, data = m2a_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m2a_data$DRXTCAFF)
pred_plus50 <- predict(model_2a, newdata = transform(m2a_data[1,], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_2a, newdata = transform(m2a_data[1,], DRXTCAFF = mean_caff))
effect_plus50_2a <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m2_cerad <- subset(df, !is.na(Y_lowCERAD) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR))
fit_m2_cerad <- glm(Y_lowCERAD ~ caff_q + RIDAGEYR + RIAGENDR, family = binomial(), data = m2_cerad)

# odds ratio
OR_m2a <- exp(cbind(OR = coef(fit_m2_cerad), confint(fit_m2_cerad)))

# saving odds ratio from fourth quartile
OR_2a <- OR_m2a["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_2a <- summary(fit_m2_cerad)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_2a <- p_2a < 0.05


# b) DSST

# linear regression DSST - Caffiene
m2b_data <- subset(df, !is.na(CFDDS) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR))
model_2b <- lm(CFDDS ~ DRXTCAFF + RIDAGEYR + RIAGENDR, data = m2b_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m2b_data$DRXTCAFF)
pred_plus50 <- predict(model_2b, newdata = transform(m2b_data[1,], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_2b, newdata = transform(m2b_data[1,], DRXTCAFF = mean_caff))
effect_plus50_2b <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m2_dsst <- subset(df, !is.na(Y_lowDSST) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR))
fit_m2_dsst <- glm(Y_lowDSST ~ caff_q + RIDAGEYR + RIAGENDR, family = binomial(), data = m2_dsst)

# odds ratio
OR_m2b <- exp(cbind(OR = coef(fit_m2_dsst), confint(fit_m2_dsst)))

# saving odds ratio from fourth quartile
OR_2b <- OR_m2b["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_2b <- summary(fit_m2_dsst)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_2b <- p_2b < 0.05


# c) Animal Fluency

# linear regression Animal Fluency - Caffiene
m2c_data <- subset(df, !is.na(CFDAST) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR))
model_2c <- lm(CFDAST ~ DRXTCAFF + RIDAGEYR + RIAGENDR, data = m2c_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m2c_data$DRXTCAFF)
pred_plus50 <- predict(model_2c, newdata = transform(m2c_data[1,], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_2c, newdata = transform(m2c_data[1,], DRXTCAFF = mean_caff))
effect_plus50_2c <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m2_af <- subset(df, !is.na(Y_lowAF) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR))
fit_m2_af <- glm(Y_lowAF ~ caff_q + RIDAGEYR + RIAGENDR, family = binomial(), data = m2_af)

# odds ratio
OR_m2c <- exp(cbind(OR = coef(fit_m2_af), confint(fit_m2_af)))

# saving odds ratio from fourth quartile
OR_2c <- OR_m2c["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_2c <- summary(fit_m2_af)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_2c <- p_2c < 0.05
```

```python
%%R
# summary of model 2
model2_results <- data.frame(
  Outcome = c("CERAD delayed recall", "DSST", "Animal Fluency"),
  OR_Q4_vs_Q1 = c(OR_2a, OR_2b, OR_2c),
  P_value_Q4_vs_Q1 = c(p_2a, p_2b, p_2c),
  Significant_0.05 = c(sig_2a, sig_2b, sig_2c),
  Effect_plus_50mg = c(effect_plus50_2a, effect_plus50_2b, effect_plus50_2c)
)
model2_results
```

### <b> 4. Model 3 (Adjusted for Further Covariates) <a name="m3"></a> </b>

Model 3: Fully adjusted association between caffeine intake and cognitive performance

Model 3 further adjusts the association between caffeine intake and cognitive performance for a set of health and lifestyle covariates in addition to age and sex. Specifically, we adjust for body mass index (BMI), race/ethnicity, marital status, smoking status, alcohol intake, diabetes, and history of stroke, as shown in Figure 3.

These variables may be associated with both caffeine consumption patterns and cognitive performance and are therefore included to reduce residual confounding. As in previous models, separate linear regressions are estimated for each cognitive outcome, and results are summarised using a 50 mg/day increase in caffeine intake.

```python
%%R
# linear regression CERAD - Caffiene
m3a_data <- subset(df, !is.na(CFDCSR) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(BMXBMI) & !is.na(RIDRETH1) & !is.na(DMDMARTL) & !is.na(SMQ020) & !is.na(ALQ101) & !is.na(DIQ010) & !is.na(MCQ160F))
model_3a <- lm(CFDCSR ~ DRXTCAFF + RIDAGEYR + RIAGENDR + BMXBMI + RIDRETH1 + DMDMARTL + SMQ020 + ALQ101 + DIQ010 + MCQ160F, data = m3a_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m3a_data$DRXTCAFF)
pred_plus50 <- predict(model_3a, newdata = transform(m3a_data[1,], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_3a, newdata = transform(m3a_data[1,], DRXTCAFF = mean_caff))
effect_plus50_3a <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m3_cerad <- subset(df, !is.na(Y_lowCERAD) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(BMXBMI) & !is.na(RIDRETH1) & !is.na(DMDMARTL) & !is.na(SMQ020) & !is.na(ALQ101) & !is.na(DIQ010) & !is.na(MCQ160F))
fit_m3_cerad <- glm(Y_lowCERAD ~ caff_q + RIDAGEYR + RIAGENDR + BMXBMI + RIDRETH1 + DMDMARTL + SMQ020 + ALQ101 + DIQ010 + MCQ160F, family = binomial(), data = m3_cerad)

# odds ratio
OR_m3a <- exp(cbind(OR = coef(fit_m3_cerad), confint(fit_m3_cerad)))

# saving odds ratio from fourth quartile
OR_3a <- OR_m3a["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_3a <- summary(fit_m3_cerad)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_3a <- p_3a < 0.05


# linear regression DSST - Caffiene
m3b_data <- subset(df, !is.na(CFDDS) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(BMXBMI) & !is.na(RIDRETH1) & !is.na(DMDMARTL) & !is.na(SMQ020) & !is.na(ALQ101) & !is.na(DIQ010) & !is.na(MCQ160F))
model_3b <- lm(CFDDS ~ DRXTCAFF + RIDAGEYR + RIAGENDR + BMXBMI + RIDRETH1 + DMDMARTL + SMQ020 + ALQ101 + DIQ010 + MCQ160F, data = m3b_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m3b_data$DRXTCAFF)
pred_plus50 <- predict(model_3b, newdata = transform(m3b_data[1,], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_3b, newdata = transform(m3b_data[1,], DRXTCAFF = mean_caff))
effect_plus50_3b <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m3_dsst <- subset(df, !is.na(Y_lowDSST) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(BMXBMI) & !is.na(RIDRETH1) & !is.na(DMDMARTL) & !is.na(SMQ020) & !is.na(ALQ101) & !is.na(DIQ010) & !is.na(MCQ160F))
fit_m3_dsst <- glm(Y_lowDSST ~ caff_q + RIDAGEYR + RIAGENDR + BMXBMI + RIDRETH1 + DMDMARTL + SMQ020 + ALQ101 + DIQ010 + MCQ160F, family = binomial(), data = m3_dsst)

# odds ratio
OR_m3b <- exp(cbind(OR = coef(fit_m3_dsst), confint(fit_m3_dsst)))

# saving odds ratio from fourth quartile
OR_3b <- OR_m3b["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_3b <- summary(fit_m3_dsst)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_3b <- p_3b < 0.05


# linear regression Animal Fluency - Caffiene
m3c_data <- subset(df, !is.na(CFDAST) & !is.na(DRXTCAFF) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(BMXBMI) & !is.na(RIDRETH1) & !is.na(DMDMARTL) & !is.na(SMQ020) & !is.na(ALQ101) & !is.na(DIQ010) & !is.na(MCQ160F))
model_3c <- lm(CFDAST ~ DRXTCAFF + RIDAGEYR + RIAGENDR + BMXBMI + RIDRETH1 + DMDMARTL + SMQ020 + ALQ101 + DIQ010 + MCQ160F, data = m3c_data)

# Estimating Causal Effect of +50 mg daily
mean_caff <- mean(m3c_data$DRXTCAFF)
pred_plus50 <- predict(model_3c, newdata = transform(m3c_data[1,], DRXTCAFF = mean_caff + 50))
pred_base   <- predict(model_3c, newdata = transform(m3c_data[1,], DRXTCAFF = mean_caff))
effect_plus50_3c <- pred_plus50 - pred_base

# logistic regression in quartiles to compare with the original paper
m3_af <- subset(df, !is.na(Y_lowAF) & !is.na(caff_q) & !is.na(RIDAGEYR) & !is.na(RIAGENDR) & !is.na(BMXBMI) & !is.na(RIDRETH1) & !is.na(DMDMARTL) & !is.na(SMQ020) & !is.na(ALQ101) & !is.na(DIQ010) & !is.na(MCQ160F))
fit_m3_af <- glm(Y_lowAF ~ caff_q + RIDAGEYR + RIAGENDR + BMXBMI + RIDRETH1 + DMDMARTL + SMQ020 + ALQ101 + DIQ010 + MCQ160F, family = binomial(), data = m3_af)

# odds ratio
OR_m3c <- exp(cbind(OR = coef(fit_m3_af), confint(fit_m3_af)))

# saving odds ratio from fourth quartile
OR_3c <- OR_m3c["caff_qQ4", "OR"]

# saving p-value from fourth quartile
p_3c <- summary(fit_m3_af)$coefficients["caff_qQ4", "Pr(>|z|)"]
sig_3c <- p_3c < 0.05
```

Model 3 results (fully adjusted)
Model 3: Fully adjusted association between caffeine intake and cognitive performance

Model 3 presents fully adjusted linear regression models examining the association between caffeine intake and cognitive performance. In addition to age and sex, models adjust for body mass index (BMI), race/ethnicity, marital status, smoking status, alcohol intake, diabetes, and history of stroke, as specified in the causal diagram for Model 3.

Separate models were estimated for each cognitive outcome. Results are summarised using a model-based contrast corresponding to a 50 mg/day increase in caffeine intake.

Model 3a: CERAD delayed recall

After full adjustment for demographic, lifestyle, and health-related covariates, a 50 mg/day increase in caffeine intake was associated with an estimated increase of 0.07 points in the CERAD delayed recall score. This estimate is smaller than those observed in Models 1 and 2, indicating substantial attenuation after accounting for additional confounders.

Model 3b: Digit Symbol Substitution Test (DSST)

In the fully adjusted model for DSST performance, a 50 mg/day increase in caffeine intake was associated with an estimated increase of 0.66 points in DSST score. Although attenuated relative to earlier models, this remained the largest association among the three cognitive outcomes.

Model 3c: Animal Fluency

For Animal Fluency, a 50 mg/day increase in caffeine intake was associated with an estimated increase of 0.22 points after full adjustment. This represents a further reduction compared with the unadjusted and age- and sex-adjusted models.

Overall interpretation of Model 3

Across all three cognitive outcomes, full adjustment led to further attenuation of the estimated associations between caffeine intake and cognitive performance. This pattern suggests that the positive associations observed in unadjusted models are largely explained by demographic, lifestyle, and health-related factors. The fully adjusted results provide limited evidence of an independent association between caffeine intake and cognitive performance in this cross-sectional NHANES sample.
