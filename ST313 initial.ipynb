```python
import kagglehub

path = kagglehub.dataset_download("nguyenvy/nhanes-19882018")
print("Path to dataset files:", path)
```

```python
import pandas as pd

demo = pd.read_csv(path + "/demographics_clean.csv")
diet = pd.read_csv(path + "/dietary_clean.csv")
mort = pd.read_csv(path + "/mortality_clean.csv")
weights = pd.read_csv(path + "/weights_clean.csv")
```

```python
df = demo.merge(diet, on="SEQN", how="left") \
         .merge(mort, on="SEQN", how="left") \
         .merge(weights, on="SEQN", how="left")
```
